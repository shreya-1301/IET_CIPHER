```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data (assuming it's already read into stock_data)
# Ensure the 'Date' column is set as the index and is a datetime type
stock_data['Date'] = pd.to_datetime(stock_data['Date'])
stock_data.set_index('Date', inplace=True)

# Define a function to create subplots for each year
def plot_yearly_data_split(data, start_year, end_year):
    # Determine the number of years to plot
    num_years = end_year - start_year + 1
    num_cols = 2  # Number of columns in the grid
    num_rows = (num_years // num_cols) + (num_years % num_cols > 0)

    # Create subplots
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, num_rows * 5))
    axes = axes.flatten()  # Flatten to iterate over all subplots easily

    # Plot data for each year
    for i, year in enumerate(range(start_year, end_year + 1)):
        yearly_data = data[data.index.year == year]
        
        if not yearly_data.empty:
            sns.lineplot(ax=axes[i], x=yearly_data.index, y=yearly_data['Close'])
            axes[i].set_title(f'Stock Prices for {year}')
            axes[i].set_xlabel('Date')
            axes[i].set_ylabel('Close Price')
        else:
            axes[i].set_visible(False)  # Hide unused subplots if no data

    # Adjust layout
    plt.tight_layout()
    plt.show()

# Call the function for your desired range
plot_yearly_data_split(stock_data, 2004, 2024)
```

```python
import streamlit as st
import pandas as pd
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt

# Function to perform anomaly detection
def detect_anomalies(df, contamination=0.05):
    # Ensure the 'Close' column exists
    if 'Close' not in df.columns:
        st.error("CSV file must contain a 'Close' column")
        return None

    # Perform anomaly detection using IsolationForest
    model = IsolationForest(contamination=contamination)
    df['Anomaly'] = model.fit_predict(df[['Close']])

    # Anomalies are marked as -1, and normal points as 1
    anomalies = df[df['Anomaly'] == -1]

    return anomalies

# Function to plot the stock price with anomalies
def plot_anomalies(df, anomalies):
    plt.figure(figsize=(10, 6))
    plt.plot(df['Date'], df['Close'], label='Close Price', color='blue')
    plt.scatter(anomalies['Date'], anomalies['Close'], color='red', label='Anomalies', zorder=5)
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.title('Stock Price with Anomalies')
    plt.legend()
    plt.xticks(rotation=45)
    st.pyplot(plt)

# Streamlit app layout
def main():
    st.title('Stock Anomaly Detection')

    # File upload
    uploaded_file = st.file_uploader("Upload a CSV file", type=['csv'])

    if uploaded_file is not None:
        # Load the CSV file
        df = pd.read_csv(r"C:\Users\shrey\Downloads\train_processed.csv")

        # Display the first few rows of the uploaded data
        st.write("Data preview:", df.head())

        # Perform anomaly detection
        anomalies = detect_anomalies(df)

        if anomalies is not None:
            st.write(f"Identified {len(anomalies)} anomalies.")
            st.write(anomalies[['Date', 'Close']])  # Display anomalies (dates with anomaly)

            # Plot the anomalies
            plot_anomalies(df, anomalies)

# Run the Streamlit app
if __name__ == "__main__":
    main()
```
